### Importing relevant libraries

```python
import pandas as pd
import numpy as np
import sklearn as sk
import matplotlib as plt
```

### Dataset containing the features of songs (acousticness, danceability......)

```python
spotifydf = pd.read_csv('something.csv')
```

```python
spotifydf.head()
```

```python
spotifydf.shape
```

```python
spotifydf = spotifydf.drop(['duration_ms','explicit','mode','release_date','Unnamed: 0'],axis=1)
```

### Cleaning the spotify dataset by dropping irrelevant features above, and normalising the un-normalised features below

```python
from sklearn.preprocessing import normalize
x_array = np.array([spotifydf['tempo'],spotifydf['loudness'],spotifydf['popularity'],spotifydf['key'],spotifydf['year']])
normalized_x =normalize(x_array)
spotifydf = spotifydf.drop(['tempo','loudness','popularity','year','key'],axis=1)
spotifydf.insert(3,'tempo',normalized_x[0])
spotifydf.insert(4,'loudness',normalized_x[1])
spotifydf.insert(5,'popularity',normalized_x[2])
spotifydf.insert(6,'year',normalized_x[3])
spotifydf.insert(7,'key',normalized_x[4])
```

```python
combo = np.array(spotifydf['title']+' '+spotifydf['artist_name'])
spotifydf = spotifydf.drop(['title','artist_name'],axis=1)
spotifydf.insert(1,'name',combo)
```

```python
spotifydf = spotifydf.drop_duplicates(subset='name')
```
